### Script to move test images ending "-1.jpg" to a separate folder
Separate folder is needed for making predictions

```python
import os
from shutil import copyfile
import pandas as pd
```

```python
#load test.csv
test = pd.read_csv('/Users/Rolf/Google Drive (parent4parent.app@gmail.com)/test_images_c4_pp.csv', 
                    index_col='filename', usecols = ['filename'])
```

```python
test.head()
```

```python
#change directory
%cd /Users/Rolf/Google Drive (parent4parent.app@gmail.com)
```

```python
#create folder "profile_picture_test"
os.makedirs('./profile_picture_test')
```

```python
#copy images that are present in test dataframe to new folder
for image in test.index:
    old = './train_images/{}'.format(image)
    new = './profile_picture_test/{}'.format(image)
    copyfile(old, new)
```
